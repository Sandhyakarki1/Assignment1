Task- 3- Outlier Detection:
 Tasks:
 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
 2. Define outliers using the 1.5 × IQR rule.
 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
 in a different color.
 4. Discuss the characteristics of these outliers and their potential impact on regional averages



```python
# import numpy library.
import numpy as np
```

```python
# Filter the dataset to include only the rows where the 'Country name' is in the list of South Asian countries
Data = df[df['Country name'].isin(south_asian_countries)]
#Create a new dataframe from the filtered data
df_quartiles = pd.DataFrame(Data)
#Calculate the first and third  quartiles for the score column
Q1 = df_quartiles['score'].quantile(0.25)
Q3 = df_quartiles['score'].quantile(0.75)
#Calculate the interQuartitle Range for the score column
IQR = Q3 - Q1
Q1_GDP = df_quartiles['Log GDP per capita'].quantile(0.25)
Q3_GDP = df_quartiles['Log GDP per capita'].quantile(0.75)
IQR_GDP = Q3_GDP - Q1_GDP
#Print the calculated quartiles and IQR for 'score'.
print("1st Quartile: ", Q1)
print("3rd Quartile: ", Q3)
#Print the calculated quartiles and IQR for 'Log GDP per capita'
print("Inter Quartile Range: ", IQR)
print("1st Quartile GDP: ", Q1_GDP)
print("3rd Quartile GDP: ", Q3_GDP)
print("Inter Quartile Range GDP: ", IQR_GDP)
```

```python
#Calculate the lower bound and upper bound for 'score'
lower_bound_score=Q1-1.5*IQR
upper_bound_score=Q3+1.5*IQR
print(lower_bound_score)
print(upper_bound_score)
```

```python
#Calculate the lower and upper bound for 'Log GDP per capita.
lower_bound_gdp=Q1_GDP-1.5*IQR_GDP
upper_bound_gdp=Q3_GDP+1.5*IQR_GDP
print(lower_bound_gdp)
print(upper_bound_gdp)
```

```python
#Create a new column 'Outliers' to flag rows as outliers
df_south_asian['Outliers']=(df_south_asian['score']<lower_bound_score)|(df_south_asian['score']>upper_bound_score)|(df_south_asian['Log GDP per capita']<lower_bound_gdp)|(df_south_asian['Log GDP per capita']>upper_bound_gdp)
```

```python
#Filter outliers and non-outliers
outliers=df_south_asian[df_south_asian['Outliers']]
non_outliers=df_south_asian[~df_south_asian['Outliers']]
```

```python
#plot a scatter plot for the outliers
plt.figure(figsize=(6,4))
sns.scatterplot(data=outliers,x='Log GDP per capita',y='score',label='Outliers(Score)',color='red')
#plot a scatter plot for non outliers
sns.scatterplot(data=non_outliers,x='Log GDP per capita',y='score',label='Non-Outliers(Score)',color='green')
plt.legend()
plt.xlabel('Log GDP per capita')
plt.ylabel('Happiness Score')
plt.title('Log GDP per capita vs Happiness Score')
plt.show()
```


In the data and graph, we can see that Afghanistan stands out as an outlier among the South Asian countries.
 This is because it has the lowest Log GDP per capita score of 0.628 and a composite score of 0.3238.
  The average score for the other countries is much higher, around 1.052.
   Afghanistan's score is far below the rest, and even Bangladesh, with the next lowest score (1.122), has almost double Afghanistan's GDP score.
   This shows a big difference between Afghanistan and the other countries.

   Impact of Afghanistan as an Outlier
Afghanistan's very low scores can distort the regional averages for South Asia.
 If we include Afghanistan without any adjustments, it could bring down the overall mean happiness score
 and other related metrics for the region, giving a false impression of how well the region is doing.
To better understand the typical trends in South Asia, we may need to exclude Afghanistan or adjust its values when doing statistical analysis.
This will give a clearer picture of how the rest of the countries are performing.
